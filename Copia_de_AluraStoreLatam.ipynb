### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por tienda
ingreso_tienda_1 = tienda["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

# Mostrar resultados
print("Ingreso total Tienda 1:", ingreso_tienda_1)
print("Ingreso total Tienda 2:", ingreso_tienda_2)
print("Ingreso total Tienda 3:", ingreso_tienda_3)
print("Ingreso total Tienda 4:", ingreso_tienda_4)
```

# 2. Ventas por categoría

```python
# Tienda 1
ventas_categoria_1 = tienda["Categoría del Producto"].value_counts().reset_index()
ventas_categoria_1.columns = ["Categoría", "Ventas"]
print("Ventas por categoría - Tienda 1:\n")
print(ventas_categoria_1.to_string(index=False))
print("\n")

# Tienda 2
ventas_categoria_2 = tienda2["Categoría del Producto"].value_counts().reset_index()
ventas_categoria_2.columns = ["Categoría", "Ventas"]
print("Ventas por categoría - Tienda 2:\n")
print(ventas_categoria_2.to_string(index=False))
print("\n")

# Tienda 3
ventas_categoria_3 = tienda3["Categoría del Producto"].value_counts().reset_index()
ventas_categoria_3.columns = ["Categoría", "Ventas"]
print("Ventas por categoría - Tienda 3:\n")
print(ventas_categoria_3.to_string(index=False))
print("\n")

# Tienda 4
ventas_categoria_4 = tienda4["Categoría del Producto"].value_counts().reset_index()
ventas_categoria_4.columns = ["Categoría", "Ventas"]
print("Ventas por categoría - Tienda 4:\n")
print(ventas_categoria_4.to_string(index=False))
```

# 3. Calificación promedio de la tienda


```python
# Valoración media por tienda
media_1 = tienda["Calificación"].mean()
print(f"Valoración media - Tienda 1: {media_1:.2f}")

media_2 = tienda2["Calificación"].mean()
print(f"Valoración media - Tienda 2: {media_2:.2f}")

media_3 = tienda3["Calificación"].mean()
print(f"Valoración media - Tienda 3: {media_3:.2f}")

media_4 = tienda4["Calificación"].mean()
print(f"Valoración media - Tienda 4: {media_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
def mostrar_top_productos(tienda, numero):
    ventas = tienda["Producto"].value_counts()

    print(f"\nTienda {numero} - Productos más vendidos:")
    print(ventas.head(5).to_string())

    print(f"\nTienda {numero} - Productos menos vendidos:")
    print(ventas.tail(5).to_string())
    print("-" * 40)

# Ejecutar para cada tienda
mostrar_top_productos(tienda, 1)
mostrar_top_productos(tienda2, 2)
mostrar_top_productos(tienda3, 3)
mostrar_top_productos(tienda4, 4)
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
envio_promedio_1 = tienda["Costo de envío"].mean()
envio_promedio_2 = tienda2["Costo de envío"].mean()
envio_promedio_3 = tienda3["Costo de envío"].mean()
envio_promedio_4 = tienda4["Costo de envío"].mean()

print(f"Costo de envío promedio - Tienda 1: ${envio_promedio_1:,.2f}")
print(f"Costo de envío promedio - Tienda 2: ${envio_promedio_2:,.2f}")
print(f"Costo de envío promedio - Tienda 3: ${envio_promedio_3:,.2f}")
print(f"Costo de envío promedio - Tienda 4: ${envio_promedio_4:,.2f}")
```

Generando gráfico

Después de realizar los análisis, es hora de transformar sus resultados en visualizaciones que le ayuden a comprender mejor los patrones y los insights encontrados.

De acuerdo con funciones que haya creado, le recomendamos que genere al menos tres gráficos. Estos gráficos deben ser de diferentes tipos (como barras, líneas, dispersión, entre otros) para presentar una visión completa de los datos.

Recuerde que los gráficos deben complementar el análisis realizado, resaltando los puntos más relevantes, como los ingresos de la tienda, la distribución de categorías de productos, las opiniones de los clientes, los productos más vendidos y/o los costes de envío.

No es necesario mostrar una visualización en cada paso, pero al final del análisis debes generar los gráficos que consideres más útiles para interpretar los datos.

Consejo: Matplotlib ofrece una variedad de tipos de gráficos que son fáciles de implementar, lo que lo convierte en una excelente opción para principiantes. Utilice su creatividad para elegir los gráficos que mejor resalten las tendencias y los conocimientos de su análisis.


```python
import matplotlib.pyplot as plt

ingresos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

plt.figure(figsize=(8,5))
plt.bar(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], ingresos, color="skyblue")
plt.title("Ingresos totales por tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Añadir columna con nombre de tienda para identificar
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todos los dataframes
df_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Agrupar por tienda y categoría
ventas_categorias = df_tiendas.groupby(["Tienda", "Categoría del Producto"]).size().reset_index(name="Ventas")

# Gráfico de barras agrupado por tienda
plt.figure(figsize=(12,6))
sns.barplot(data=ventas_categorias, x="Categoría del Producto", y="Ventas", hue="Tienda")
plt.title("Cantidad de productos vendidos por categoría y tienda")
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos de ejemplo (reemplaza esto con tus DataFrames reales)
np.random.seed(42)
tienda1 = pd.DataFrame({'Número de Pedido': range(1, 51), 'Peso (kg)': np.random.uniform(0.5, 5, 50), 'Costo de envío': np.random.uniform(2000, 8000, 50)})
tienda2 = pd.DataFrame({'Número de Pedido': range(1, 76), 'Peso (kg)': np.random.uniform(1, 7, 75), 'Costo de envío': np.random.uniform(3000, 10000, 75)})
tienda3 = pd.DataFrame({'Número de Pedido': range(1, 101), 'Peso (kg)': np.random.uniform(0.2, 3, 100), 'Costo de envío': np.random.uniform(1500, 6500, 100)})
tienda4 = pd.DataFrame({'Número de Pedido': range(1, 61), 'Peso (kg)': np.random.uniform(2, 9, 60), 'Costo de envío': np.random.uniform(4000, 12000, 60)})

# Calcular el costo de envío promedio para cada tienda
tiendas_data = {
    "Tienda 1": tienda1["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

nombres_tiendas = list(tiendas_data.keys())
costos_promedio = list(tiendas_data.values())

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(costos_promedio, labels=nombres_tiendas, autopct='%1.1f%%', startangle=140)
plt.title("Proporción del Costo de Envío Promedio por Tienda")
plt.axis('equal')  # Equal aspect ratio asegura que el gráfico sea un círculo.
plt.show()
```
